#### Script to have a look at the distributions of *Silene latifolia* common garden data

```r
#clear workspace
rm(list=ls())

ls()
```

##### Load required packages

```r
library(tidyverse)
library(ggplot2)
library(cowplot)
```

##### Import dataset

```r
data <- read_csv("/Volumes/disk2/github/Silene-Invasive-Data/data/Garden_data_summed_across_censuses.csv")
```

##### Let's get the structure of the data

```r
data$Structure_region <- as.character(data$Structure_region)
str(data)
```

##### Summarize data overall, without regard to specific characters

```r
summary(data)
```

##### Let's generate a few histograms for the full dataset
